```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("social_media_survey2.csv")
```

```python
df.head(5)
```

```python
df.tail()
```

```python
df.columns
```

```python
anxiety_counts = df['Have you ever felt anxious after using social media?  '].value_counts()
anxiety_counts.plot.pie(autopct='%1.1f%%', startangle=90, shadow=True, figsize=(6,6))
plt.title("Anxiety After Using Social Media")
plt.ylabel("")
plt.show()
```

```python
hours = df['On average, how many hours per day do you spend on social media?  '].value_counts()

# Bar chart
hours.plot(kind='bar', color='skyblue')
plt.title("Daily Social Media Usage")
plt.xlabel("Hours")
plt.ylabel("Number of Responses")
plt.xticks(rotation=45)
plt.show()
```

```python
comparison_counts = df['Do you often compare yourself to others on social media (appearance, lifestyle, achievements)?  '].value_counts()

# Bar chart
comparison_counts.plot(kind='bar', color='orchid')
plt.title("Self-Comparison on Social Media")
plt.xlabel("Response")
plt.ylabel("Number of People")
plt.xticks(rotation=45)
plt.show()
```

```python
sleep_effect = df['Has social media affected your sleep pattern?  '].value_counts()

sleep_effect.plot.pie(autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'salmon'], figsize=(6,6))
plt.title("Impact of Social Media on Sleep")
plt.ylabel("")
plt.show()
```

```python
fomo = df['Do you experience FOMO (Fear of Missing Out) when seeing others’ posts about events or achievements?  '].value_counts()

fomo.plot(kind='bar', color='lightcoral')
plt.title("FOMO Due to Social Media")
plt.xlabel("Response")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
from collections import Counter

# Drop NaN first
feeling_data = df['How do you usually feel after spending a lot of time on social media? (Choose all that apply)'].dropna()

# Flatten all answers into a single list
all_feelings = []
for entry in feeling_data:
    all_feelings.extend([i.strip() for i in entry.split(',')])

# Count them
feeling_counts = Counter(all_feelings)

# Plot
plt.bar(feeling_counts.keys(), feeling_counts.values(), color='mediumslateblue')
plt.title("Feelings After Social Media Usage")
plt.xlabel("Feeling")
plt.ylabel("Frequency")
plt.xticks(rotation=30)
plt.show()
```

```python
impact_rating = df['On a scale of 1 to 5, how would you rate the overall impact of social media on your mental health? (Linear scale: 1 = Very Negative, 5 = Very Positive)']

impact_rating.plot.hist(bins=5, color='mediumaquamarine', edgecolor='black')
plt.title("Overall Mental Health Impact Rating")
plt.xlabel("Rating (1=Very Negative, 5=Very Positive)")
plt.ylabel("Number of Responses")
plt.show()
```
